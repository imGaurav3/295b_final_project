```python
import pandas as pd

# Replace 'your_file.csv' with the exact file name and extension
file_path = '/content/drive/MyDrive/final project sample data/10000 Movies Data'
data = pd.read_csv(file_path)
```

```python
data.head()
```

```python
# Assuming 'column_name' is the name of your column
unique_values = data['Genres'].unique()

# To display the unique values
print(unique_values)
```

```python
from ast import literal_eval
from pandas import json_normalize
```

```python
# Convert string representations to actual lists
data['Genres'] = data['Genres'].apply(literal_eval)

# Explode 'Genres' into separate rows
exploded_genres = data.explode('Genres')

# Normalize the dictionaries in each row to separate columns
normalized_genres = json_normalize(exploded_genres['Genres'])

# Display the result
print(normalized_genres)
```

```python
# Assuming 'name' is the column with genre names in normalized_genres
genre_counts = normalized_genres['name'].value_counts()

# Display the genre counts
print(genre_counts)
```
